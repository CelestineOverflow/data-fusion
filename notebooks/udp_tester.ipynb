```python
import socket
import time

hostname=socket.gethostname()
local_network_ip=socket.gethostbyname(hostname)
port = 1337

#send a message to the server
def send_message(message):
    sock = socket.socket(socket.AF_INET, socket.SOCK_DGRAM)
    sock.sendto(message.encode(), (local_network_ip, port))
    sock.close()

send_message("Hello World!")
```

```python
from zeroconf import ServiceInfo, Zeroconf
import socket

# get local IP address
s = socket.socket(socket.AF_INET, socket.SOCK_DGRAM)
s.connect(("8.8.8.8", 80))
local_ip = s.getsockname()[0]
s.close()

info = ServiceInfo(
    "_http._tcp.local.",
    "My Service._http._tcp.local.",
    addresses=[socket.inet_aton(local_ip)],
    port=1234,
    properties={'version': '1.0.0'},
    server="my-service.local.",
)

zeroconf = Zeroconf()
print("Registration of a service...")
zeroconf.register_service(info)
print("Registration done.")
    
```

```python
from zeroconf import ServiceBrowser, Zeroconf

#check if the service is available
#service is configure as 
info = ServiceInfo(
    # the service type name, in the form _service._proto.domain.
    "_udpservice._udp.local.",
    # the name of the service instance, in the form <Instance>.<Service>.<Domain>.
    "My Data Fusion._udpservice._udp.local.",
    addresses=[socket.inet_aton(local_ip)],
    port=1337,
    properties={'version': '1.0.0'},
    server="data-fusion.local.",
)

def on_service_state_change(zeroconf, service_type, name, state_change):
    print("service state changed")
    print("service_type: ", service_type)
    print("name: ", name)
    print("state_change: ", state_change)
    print("info: ", info)
    print("info.name: ", info.name)
    print("info.type: ", info.type)
    print("info.addresses: ", info.addresses)
    print("info.port: ", info.port)
    print("info.properties: ", info.properties)
    print("info.server: ", info.server)
print("finding service...")
zeroconf = Zeroconf()
browser = ServiceBrowser(zeroconf, "_udpservice._udp.local.", handlers=[on_service_state_change])
print("service found")
    
```
